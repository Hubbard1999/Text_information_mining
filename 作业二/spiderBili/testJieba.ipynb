```python
import jieba

data_path = "danmu.txt"
data_stop_path = "baidu_stopwords.txt"

data_list = open(data_path, encoding='utf-8').readlines()
data_list = [line.strip("\n") for line in data_list]
# data_list

data_stop_list = open(data_stop_path, encoding='utf-8').readlines()
data_stop_list = [line.strip("\n") for line in data_stop_list]
data_stop_list
```

```python
import jieba
word_list=set()
for item in data_list:
    seg_list = jieba.lcut(item)

    for seg_item in seg_list:
        if seg_item in data_stop_list:
            continue
        word_list.add(seg_item)
word_list
```

```python
for i in word_list:
    with open('fenci.txt', 'a',encoding='utf-8') as f:
        f.write(i + '\n')
```

```python
import jieba
jieba.load_userdict('bilibili.txt')
word_list=[]
for item in data_list:
    seg_item = " ".join(jieba.lcut(item))
    word_list.append(seg_item)
word_list
```

```python
import jieba
word_list={}
word_sum = 0
for item in data_list:
    seg_list = jieba.lcut(item)

    for seg_item in seg_list:

        if seg_item in data_stop_list:
            continue
        if seg_item in word_list.keys():
            word_list[seg_item] += 1
        else:
            word_list[seg_item] = 1
        word_sum+=1
```

```python
word_list = sorted(word_list.items(), key=lambda x: x[1], reverse=True)
word_sum
word_list
```

```python
word_tf = {}
for key,value in word_list:
    word_tf[key] = value/word_sum
word_tf
```

```python
new_word = {k:v for k, v in word_tf.items() if v>=0.0005}
len(new_word)
```

```python
for k, _ in new_word.items():
    with open('bilibili.txt', 'a',encoding='utf-8') as f:
        f.write(k + '\n')
```
